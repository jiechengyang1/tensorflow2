```python
import tensorflow as tf
import numpy as np
```

```python
w = tf.Variable(tf.constant(5, dtype=tf.float32))
lr = 0.2
epoch = 40
for epoch in range(epoch):  # for epoch epoch1w,constant540
    with tf.GradientTape() as tape:  # withgrads
        loss = tf.square(w + 1)
    grads = tape.gradient(loss, w)  # .gradient

    w.assign_sub(lr * grads)  # .assign_sub  w -= lr*grads  w = w - lr*grads
    print("After %s epoch,w is %f,loss is %f" % (epoch, w.numpy(), loss))

# lr0.2     0.001  0.999 
#  loss   w = -1 w 
```

```python
import tensorflow as tf
from sklearn import datasets
from matplotlib import pyplot as plt
import numpy as np
```

```python
#
x_data = datasets.load_iris().data
y_data = datasets.load_iris().target
print(x_data)
print(y_data)
```

```python
# 
np.random.seed(116)  # seed
np.random.shuffle(x_data)
np.random.seed(116)
np.random.shuffle(y_data)
tf.random.set_seed(116)
```

```python
#15012030
x_train = x_data[:-30]
y_train = y_data[:-30]
x_test = x_data[-30:]
y_test = y_data[-30:]
```

```python
#
x_train = tf.cast(x_train, tf.float32)
x_test = tf.cast(x_test, tf.float32)
```

```python
# from_tensor_slicesbatch
train_db = tf.data.Dataset.from_tensor_slices((x_train, y_train)).batch(32)
test_db = tf.data.Dataset.from_tensor_slices((x_test, y_test)).batch(32)
```

```python
#
w1=tf.Variable(tf.random.truncated_normal([4,3],stddev=0.1,seed=1))
b1=tf.Variable(tf.random.truncated_normal([3],stddev=0.1,seed=1))
```

```python
lr = 0.1  # 0.1
train_loss_results = []  # lossloss
test_acc = []  # accacc
epoch = 500  # 500
loss_all = 0  # 4steploss_allstep4loss

# 
for epoch in range(epoch):  #epoch
    for step, (x_train, y_train) in enumerate(train_db):  #batch stepbatch
        with tf.GradientTape() as tape:  # with
            y = tf.matmul(x_train, w1) + b1  # 
            y = tf.nn.softmax(y)  # yloss
            y_ = tf.one_hot(y_train, depth=3)  # lossaccuracy
            loss = tf.reduce_mean(tf.square(y_ - y))  # mse = mean(sum(y-out)^2)
            loss_all += loss.numpy()  # steplosslossloss
        # loss
        grads = tape.gradient(loss, [w1, b1])

        #  w1 = w1 - lr * w1_grad    b = b - lr * b_grad
        w1.assign_sub(lr * grads[0])  # w1
        b1.assign_sub(lr * grads[1])  # b

    # epochloss
    print("Epoch {}, loss: {}".format(epoch, loss_all/4))
    train_loss_results.append(loss_all / 4)  # 4steploss
    loss_all = 0  # loss_allepochloss
```

```python
    # 
    # total_correct, total_number0
    total_correct, total_number = 0, 0
    for x_test, y_test in test_db:
        # 
        y = tf.matmul(x_test, w1) + b1
        y = tf.nn.softmax(y)
        pred = tf.argmax(y, axis=1)  # y
        # predy_test
        pred = tf.cast(pred, dtype=y_test.dtype)
        # correct=10boolint
        correct = tf.cast(tf.equal(pred, y_test), dtype=tf.int32)
        # batchcorrect
        correct = tf.reduce_sum(correct)
        # batchcorrect
        total_correct += int(correct)
        # total_numberx_testshape[0]
        total_number += x_test.shape[0]
    # total_correct/total_number
    acc = total_correct / total_number
    test_acc.append(acc)
    print("Test_acc:", acc)
    print("--------------------------")
```

```python
#  loss 
plt.title('Loss Function Curve')  # 
plt.xlabel('Epoch')  # x
plt.ylabel('Loss')  # y
plt.plot(train_loss_results, label="$Loss$")  # trian_loss_resultsLoss
plt.legend()  # 
plt.show()  # 

```

```python
#  Accuracy 
plt.title('Acc Curve')  # 
plt.xlabel('Epoch')  # x
plt.ylabel('Acc')  # y
plt.plot(test_acc, label="$Accuracy$")  # test_accAccuracy
plt.legend()
plt.show()
```
